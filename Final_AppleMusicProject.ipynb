```python
#Playlist vs Album Source Share

def categorize_origin(origin):
    if 'PLAYLIST' in origin:
        return 'Playlist'
    elif 'ALBUM' in origin:
        return 'Album'
    elif 'RADIO' in origin:
        return 'Radio'
    elif 'MIX' in origin:
        return 'Mix'
    else:
        return 'Other'

playlist_source_df['origin_category'] = playlist_source_df['origin'].astype(str).apply(categorize_origin)
source_counts = playlist_source_df['origin_category'].value_counts().reset_index()
source_counts.columns = ['source_type', 'play_count']
source_counts['percent_share'] = (source_counts['play_count'] / source_counts['play_count'].sum() * 100).round(2)
print(source_counts)
```

```python
import pandas as pd
import glob
import os
from google.colab import files


path = '/content/drive/MyDrive/Drive_Mount '
all_files = sorted(glob.glob(os.path.join(path, 'play_activity_part_*.csv')))
df_list = [pd.read_csv(f, low_memory=False) for f in all_files]
df = pd.concat(df_list, ignore_index=True)

timestamp_col = None
for col in df.columns:
    if 'timestamp' in col.lower() or 'time' in col.lower():
        timestamp_col = col
        break

if not timestamp_col:
    raise ValueError(" Could not find a time-related column.")

df[timestamp_col] = pd.to_datetime(df[timestamp_col], errors='coerce')
df['hour'] = df[timestamp_col].dt.hour

def get_time_block(hour):
    if 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 22:
        return 'Evening'
    else:
        return 'Late Night'

df['time_of_day'] = df['hour'].apply(get_time_block)

track_col = None
for col in df.columns:
    if 'track' in col.lower() or 'song' in col.lower():
        track_col = col
        break

if not track_col:
    raise ValueError(" Could not find a track name column.")

recommendations = (
    df.groupby(['time_of_day', track_col])
      .size()
      .reset_index(name='play_count')
      .sort_values(['time_of_day', 'play_count'], ascending=[True, False])
      .groupby('time_of_day')
      .head(5)
)

output_file = 'time_of_day_recommendations.csv'
recommendations.to_csv(output_file, index=False)
files.download(output_file)

print(" Done! Downloading time-of-day recommendations.")
```

```python
import pandas as pd
import glob
import os
from google.colab import files

path = '/content/drive/MyDrive/Drive_Mount '
all_files = sorted(glob.glob(os.path.join(path, 'play_activity_part_*.csv')))

if not all_files:
    raise ValueError(" No play_activity_part files found in Drive_Mount.")

#Read and merge
df_list = [pd.read_csv(f, low_memory=False) for f in all_files]
df = pd.concat(df_list, ignore_index=True)

timestamp_col = None
for col in df.columns:
    if 'timestamp' in col.lower() or 'time' in col.lower():
        timestamp_col = col
        break

if not timestamp_col:
    raise ValueError(" Could not find a time-related column.")

df[timestamp_col] = pd.to_datetime(df[timestamp_col], errors='coerce')
df['hour'] = df[timestamp_col].dt.hour
df['weekday'] = df[timestamp_col].dt.day_name()

def get_time_block(hour):
    if 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 22:
        return 'Evening'
    else:
        return 'Late Night'

df['time_of_day'] = df['hour'].apply(get_time_block)

heatmap = (
    df.groupby(['weekday', 'time_of_day'])
    .size()
    .reset_index(name='play_count')
    .sort_values(by=['weekday', 'time_of_day'])
)

output_file = 'weekday_hour_heatmap.csv'
heatmap.to_csv(output_file, index=False)
files.download(output_file)

print(" Weekday vs Time-of-Day heatmap data generated and downloaded.")
```

```python
import pandas as pd
import sqlite3
import glob
import os
from google.colab import files

path = '/content/drive/MyDrive/Drive_Mount '
all_files = sorted(glob.glob(os.path.join(path, 'play_activity_part_*.csv')))
df_list = [pd.read_csv(file, low_memory=False) for file in all_files]
df = pd.concat(df_list, ignore_index=True)

df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace(r'[^\w_]', '', regex=True)

timestamp_col = None
for col in df.columns:
    if 'time' in col.lower():
        timestamp_col = col
        break

if timestamp_col is None:
    raise ValueError(" No time-related column found in part files.")

df[timestamp_col] = pd.to_datetime(df[timestamp_col], errors='coerce')
df = df.dropna(subset=[timestamp_col])
df['hour'] = df[timestamp_col].dt.hour
df['weekday'] = df[timestamp_col].dt.day_name()

conn = sqlite3.connect(':memory:')
df.to_sql('plays', conn, index=False, if_exists='replace')
print(f" Timestamp column used: {timestamp_col}")
print(" Data loaded into SQLite.")
```

```python
import pandas as pd

query = """
SELECT
    strftime('%w', event_end_timestamp) AS weekday_num,
    strftime('%Y-%m-%d', event_end_timestamp) AS date,
    COUNT(*) AS play_count
FROM plays
WHERE event_end_timestamp IS NOT NULL
GROUP BY weekday_num, date
ORDER BY date;
"""

weekday_summary = pd.read_sql_query(query, conn)

weekday_map = {
    '0': 'Sunday', '1': 'Monday', '2': 'Tuesday', '3': 'Wednesday',
    '4': 'Thursday', '5': 'Friday', '6': 'Saturday'
}
weekday_summary['weekday'] = weekday_summary['weekday_num'].map(weekday_map)

output_file = 'sql_weekday_summary.csv'
weekday_summary.to_csv(output_file, index=False)

from google.colab import files
files.download(output_file)

print(" SQL weekday summary generated and downloaded.")
```

```python
weekday_totals = (
    weekday_summary
    .groupby('weekday', sort=False)['play_count']
    .sum()
    .reset_index()
    .sort_values(by='play_count', ascending=False)
)

weekday_totals.to_csv("sql_weekday_totals.csv", index=False)
files.download("sql_weekday_totals.csv")
```
